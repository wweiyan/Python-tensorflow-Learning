```python
%matplotlib inline
import sys
import os
import cv2
import dlib
from matplotlib import pyplot as plt

img = "Aaron_Eckhart_0001.jpg"

detector = dlib.get_frontal_face_detector()
img_a = cv2.imread(img)
gray_img = cv2.cvtColor(img_a,cv2.COLOR_RGB2GRAY)#图片灰度化
```

```python
plt.imshow(gray_img)
plt.show()
```

```python
#dlib 取出人脸特征值，并对原图片进行切割压缩
dets = detector(gray_img,1)
for i,d in enumerate(dets):
    x1 = d.top() if d.top() > 0 else 0
    y1 = d.bottom() if d.bottom() > 0 else 0
    x2 = d.left() if d.left() > 0 else 0
    y2 = d.right() if d.right() > 0 else 0

face = gray_img[x1:y1,x2:y2]#切割图片
face = cv2.resize(face,(64,64))#压缩图片
plt.imshow(face)
plt.show()
```
